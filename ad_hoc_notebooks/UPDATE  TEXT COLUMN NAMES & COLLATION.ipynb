# FUNCTIONS 

```python
def query_for_all_tables_in_db(connection_string, database):
    engine=create_engine(connection_string)
    table_data_raw=pd.read_sql_query(sql=f"""
    select TABLE_NAME
    from information_schema.tables
    where TABLE_SCHEMA = '{database}' 
    """, con=engine)
    return table_data_raw
```

```python
def get_data_type(connection_string, table_name, COLUMN_NAME, db):
    COLUMN_NAME = COLUMN_NAME.replace('`',"")
    engine=create_engine(connection_string)
    q = f"""
SELECT COLUMN_TYPE FROM INFORMATION_SCHEMA.COLUMNS 
WHERE table_name = '{table_name}' AND COLUMN_NAME = '{COLUMN_NAME}' and table_schema = '{db}';"""
    table_data_raw=pd.read_sql_query(sql=q, con=engine)
    print(q)
    return table_data_raw.iloc[0]['COLUMN_TYPE']
```

```python
def test_collation(connection_string, tables_list):
    engine=create_engine(connection_string)
    tables_good = []
    tables_bad = []
    for t in tables_list:
        table_data_raw=pd.read_sql_query(sql=f"""
select count(*)
from {t} a
where a.text != CONVERT(a.text USING utf8mb4)
        """, con=engine)
        print(table_data_raw)
        if table_data_raw.iloc[0]['count(*)'] !=0:
            tables_bad.append(t)
        else:
            tables_good.append(t)
            
    return tables_bad, tables_good
```
